## FibsemServer Client 

```python
%load_ext autoreload
%autoreload 2


from fibsem.server.client import FibsemClient
client = FibsemClient()
```

```python
from fibsem.structures import BeamType
sem_image = client.acquire_image(beam_type=BeamType.ELECTRON)
fib_image = client.acquire_image(beam_type=BeamType.ION)
```

```python
import math
client.move_to_milling_angle(math.radians(15))
```

```python
from fibsem.structures import FibsemRectangleSettings

rect = FibsemRectangleSettings(20e-6, 10e-6, 1e-6, 0, 0)
client.draw_patterns([rect])
```

```python
client.clear_patterns()
```

```python
from fibsem.milling.base import FibsemMillingSettings
client.setup_milling(FibsemMillingSettings(milling_current=0.1e-9))
```

```python
from fibsem.structures import FibsemRectangleSettings

rect = FibsemRectangleSettings(20e-6, 10e-6, 0.1e-6, 0, 0)
client.draw_patterns([rect])
```

```python
client.run_milling(1e-9, 30e3)
```

```python
client.stop_milling()
```

```python
client.finish_milling(40e-12, 30e3)
```

```python
client.acquire_image(BeamType.ELECTRON)
client.acquire_image(BeamType.ION)
```

```python
client.get_current_milling_angle()
```

```python
from fibsem.structures import FibsemStagePosition
client.move_stage_relative(FibsemStagePosition(x=-20e-6))

client.acquire_image(BeamType.ELECTRON)
client.acquire_image(BeamType.ION)
```

```python
from fibsem import acquire
from fibsem.structures import ImageSettings

image_settings = ImageSettings(hfw=500e-6)
client.set_field_of_view(500e-6, BeamType.ELECTRON)
acquire.acquire_channels(client,image_settings)
```

```python
image_settings.beam_type = BeamType.ELECTRON
image_settings.field_of_view = 200e-6
client.acquire_image(image_settings=image_settings)
```

```python
from fibsem.milling.tasks import FibsemMillingTaskConfig, run_milling_task, FibsemMillingStage

config = FibsemMillingTaskConfig(stages=[FibsemMillingStage()])
config.stages[0].milling.milling_current = 1e-9


run_milling_task(client, config)
```
